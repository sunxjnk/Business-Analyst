```python
import pandas as pd
import numpy as np
```

```python
#câu 1: Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
df = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')
print("số dòng là: ", df.shape[0])
print("số cột là: ", df.shape[1])
print("Kiểu dữ liệu là: ", df.dtypes)
```

```python
#Câu 2: 
"""Tạo cột mới có tên quý - ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm
trong các tháng 1,2,3; nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng
4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9; nhận giá trị
4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;"""
#Tách ngày, tháng, năm
df['Day']=pd.to_datetime(df['InvoiceDate']).dt.day
df['Month']=pd.to_datetime(df['InvoiceDate']).dt.month
df['Year']=pd.to_datetime(df['InvoiceDate']).dt.year
df
```

```python
#Điều kiện
cond = [(df['Month'] < 4 ), (df['Month'] > 3) & (df['Month'] < 7) , (df['Month'] > 7) & (df['Month'] < 10), (df['Month'] > 10) & (df['Month'] < 13)]
choices = ['I','II','III','IV']
```

```python
#thêm quý dựa trên điều kiện
df['Quarter'] = np.select(cond, choices,default=np.nan )
df
```

```python
#Câu 3: 
df['Amount'] = df['Quantity'] * df['UnitPrice']
df
```

```python
#Câu 4: Tạo cột mới ‘Discount’ 
"""nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 
5% nếu là ‘France’ ngược lại là 0%. """
condition = [(df['Country'] == 'United Kingdom' ) & (df['Quarter'] == 'IV') , (df['Country'] == 'France')]
choice = ["10%","5%"]
```

```python
#thêm Discount dựa trên điều kiện
df['Discount'] = np.select(condition, choice, default = "0%" )
df
```

```python
#Câu 5: Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.
df['Total'] = df['Amount'] - df['Discount']
df
```
