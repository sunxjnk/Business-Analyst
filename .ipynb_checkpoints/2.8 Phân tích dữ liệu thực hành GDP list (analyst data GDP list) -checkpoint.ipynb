```python
import pandas as pd
```

```python
#Câu 1: Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
df = pd.read_csv('GDPlist.csv', encoding = 'ISO-8859-1')
```

```python
#câu 1:Cho biết số dòng, số cột
print("số dòng là: ", df.shape[0])
print("số cột là: ", df.shape[1])
```

```python
#câu 1: Kiểu dữ liệu các thuộc tính
df.dtypes
```

```python
#câu 2: Việt hóa tên các cột trong bảng dữ liệu: Country �� Nuoc; Continent �� Chauluc; GDP (millions of US$) �� GDP (trieu $)
df.rename(columns={'Country':'Quocgia','Continent':'Chauluc','GDP (millions of US$)':'GDP(trieu $)'},inplace=True)
```

```python
df.head(10)
```

```python
#Câu 3: Chèn thêm một cột “Thanhpho” vào sau cột “Nuoc”, giá trị ban đầu là giá trị của cột “Nuoc”
df.insert(1,'City', df.QuocGia)
```

```python
#Câu 5:  Xóa các bản ghi có Chauluc là ‘Asia’
df.drop(df.loc[df['Chauluc'] == 'Asia'].index, axis = 0, inplace = True)
df
```

```python
#Câu 6: Xóa các bản ghi có GDP < 300000
df.drop(df.loc[df['GDP (trieu $)'] < 30000].index, axis = 0, inplace = True)
df
```
