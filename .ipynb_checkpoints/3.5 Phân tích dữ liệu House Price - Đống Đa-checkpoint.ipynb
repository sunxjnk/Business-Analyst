```python
import pandas as pd
```

```python
#Câu 1: Đọc bộ dữ liệu
df = pd.read_excel('house_price_đống đa.xlsx')
df
```

```python
#Cho biết số dòng, số cột
print("số dòng là: ", df.shape[0])
print("số cột là: ", df.shape[1])
```

```python
#Kiểu dữ liệu các thuộc tính
df.dtypes
```

```python
#Câu 2: Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên
df.query("(ward_name == ['Phường Trung Liệt','Phường Khâm Thiên']) and (type_of_land == 'Bán nhà riêng')")
```

```python
#Câu 3: Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.
df.query("(land_certificate == ['Sổ đỏ']) and (bedroom >= 4)")[['address','price','house_direction','balcony_direction']]
```

```python
#Câu 4: Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.
avg = df.groupby(['type_of_land'])['price'].mean().to_frame()
avg.rename(columns={'price':'price_avg'}, inplace = True)
minp = df.groupby(['type_of_land'])['price'].min().to_frame()
minp.rename(columns={'price':'price_min'}, inplace = True)
maxp = df.groupby(['type_of_land'])['price'].max().to_frame()
maxp.rename(columns={'price':'price_max'}, inplace = True)
df4 = pd.concat([avg, minp, maxp], axis = 1)
df4
```

```python
#Câu 5: Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường.
br = df.groupby(['ward_name'])['bedroom'].mean().to_frame()
vs = df.groupby(['ward_name'])['toilet'].mean().to_frame()
fl = df.groupby(['ward_name'])['floor'].mean().to_frame()
df5 = pd.concat([br, vs, fl], axis = 1)
df5
```
