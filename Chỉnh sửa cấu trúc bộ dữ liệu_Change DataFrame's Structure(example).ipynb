```python
import pandas as pd
```

```python
df=pd.read_csv('FoodPrice_in_Turkey.csv')
df.head()
```

```python
#change name of columns
df.rename(columns={'Place':'Địa điểm','ProductName':'Tên SP'},inplace=True)
df.head()
```

```python
#add a new column with value NaN
df['Note'] = 'NaN'
df.head()
```

```python
#add a sale column with value 10%
df['Giảm giá']= pd.Series('10%', index=df.index)
df.head()
```

```python
df=df.append({'Địa điểm':'NA','ProductId':'RR','Tên SP':'Rice','UmId':10,'UmName':'KG'}, index = True)
df.tail()
```

```python
#delete a column
del df['Note']
df.head()
```

```python
#delete a column and show what it is
df.pop('Giảm giá')
```

```python
#delete mutiples columns
df.drop(['Month','Year'], axis=1, inplace=True)
df
```

```python
#delete the row number n. In this example, we delete the 1st row
df.drop(1, axis = 0, inplace=True)
df
```
